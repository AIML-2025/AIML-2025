```python
import pandas as pd

# List of salaries
salaries = [5000, 5575, 6100, 6555, 7000, 7575, 8000, 8500, 9000, 9200, 10000, 10500, 11200, 11500, 12000,5000]

# Convert to DataFrame
df = pd.DataFrame(salaries, columns=['Salary'])
df['Employee_ID'] = range(1, len(df) + 1)
```

```python
print("First 5 rows:\n", df.head())
print("Last 5 rows:\n", df.tail())
```

```python
print("Mean:", df['Salary'].mean())
print("Median:", df['Salary'].median())
print("Mode:", df['Salary'].mode()[0])
print("Standard Deviation:", df['Salary'].std())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.histplot(df['Salary'])
plt.title('Distribution of Salaries')
plt.xlabel('Salaries')
plt.ylabel('Frequency')
plt.xticks(np.arange(5000, 13000, 1000))
plt.show()
```

```python
sns.boxplot(x=df['Salary'], color='orange')
plt.title('Box Plot of Salaries')
plt.xticks(np.arange(0, 130, 10))
plt.show()
```

```python
salary_counts = df['Salary'].value_counts()
plt.pie(salary_counts, labels=salary_counts.index, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Salary Distribution')
plt.show()
```

```python
sns.kdeplot(df['Salary'], fill=True, color='blue')
plt.title('KDE Plot of Salaries')
plt.xlabel('Salaries')
plt.ylabel('Density')
plt.show()
```

```python
import statsmodels.api as sm

sm.qqplot(df['Salary'], line='s')
plt.title('Q-Q Plot of Salaries')
plt.show()
```

```python
# Calculate average salary
average_salary = df['Salary'].mean()

# Calculate maximum salary
max_salary = df['Salary'].max()

# Find employee with the highest salary
employee_high_salary = df.loc[df['Salary'] == max_salary]

# Find employee with the lowest salary
min_salary = df['Salary'].min()
employee_low_salary = df.loc[df['Salary'] == min_salary]

# Display results
print("Average Salary:", average_salary)
print("Maximum Salary:", max_salary)
print("Employee with Highest Salary:\n", employee_high_salary)
print("Employee with Lowest Salary:\n", employee_low_salary)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# List of salaries
salaries = [5000, 5575, 6100, 6555, 7000, 7575, 8000, 8500, 9000, 9200, 10000, 10500, 11200, 11500, 12000, 5000]

# Convert to DataFrame
df = pd.DataFrame(salaries, columns=['Salary'])
df['Employee_ID'] = range(1, len(df) + 1)

# Sort the DataFrame by Salary in descending order
sorted_df = df.sort_values(by='Salary', ascending=False)

# Define a list of custom colors (one for each bar)
colors = sns.color_palette("coolwarm", len(sorted_df))

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Employee_ID', y='Salary', data=sorted_df, hue='Employee_ID', dodge=False, palette=colors)

# Add titles and labels
plt.title('Salaries from High to Low')
plt.xlabel('Employee ID')
plt.ylabel('Salary')
plt.xticks(sorted_df['Employee_ID'])  # Set x-ticks to Employee IDs

# Adjust legend position
plt.legend(title='Employee ID', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```
